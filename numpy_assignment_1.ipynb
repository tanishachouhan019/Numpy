# Q1. Create a NumPy Array

## Problem Statement
Create a NumPy array using the list:
```python
[10, 20, 30, 40, 50]
```

## Sample Output
```python
[10 20 30 40 50]
```


````python
import numpy as np
l = [10,20,30,40,50]
array = np.array(l)
print(array)
````
